```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import math
data_raw = pd.read_csv(
    filepath_or_buffer='https://raw.githubusercontent.com/Sabaae/Dataset/main/TemperaturesbyCity.csv',
    index_col=0
)
```

```python
#1
def plot_time_series(df, city_list):

  num_city=len(city_list)

  fig, axes = plt.subplots(nrows=num_city, figsize=(10, 4*num_city))

  for i in range(num_city):
    row=df.loc[city_list[i]]
    row.plot(ax=axes[i])



```

```python
#2
col_names=data_raw.columns
row_names=data_raw.index

scaler=StandardScaler().fit(data_raw)
data_scaled=scaler.transform(data_raw)
print(data_scaled)
df_scaled=pd.DataFrame(data_scaled, index=row_names, columns=col_names)
df_scaled
```

```python
#3
my_city_list=['Yakeshi', 'Zamboanga', 'Norilsk', 'Juliaca', 'Doha']
plot_time_series(df_scaled, my_city_list)
```
